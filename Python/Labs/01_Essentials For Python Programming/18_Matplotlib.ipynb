## Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The code `%matplotlib inline` above is to make plots show in the notebook output.

```python
#| fig-cap: 'A Basic line plot'
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
fig = plt.figure(figsize=(6,3))
plt.plot(x, y, linewidth=1)
plt.show()
```

## Saving Plots

```python
#| fig-cap: 'Saved plot'
x = [0, 2, 4, 6]
y = [1, 2, 4, 8]
fig = plt.figure(figsize=(6,3))
plt.plot(x, y, linewidth=1)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plotted x and y values')
plt.legend(['Data 1'])

# Saving the plot 
plt.savefig('plot1.png', dpi=350, bbox_inches='tight')

plt.show()
```

## Multi Line plots

\begin{itemize}
    \item[\textcolor{cyan}{\ding{51}}] Multi Line Plots
    \item[\textcolor{cyan}{\ding{51}}] Object-Oriented Interface
\end{itemize}

```python
#| fig-cap: 'Multi plots'
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize=(6,3))
fig, ax = plt.subplots()

ax.plot(x, y)
ax.plot(x, z)

ax.xaxis.set_label_text('Angle')
ax.yaxis.set_label_text('Sine and Cosine')
ax.legend(['Sin', 'Cos'])
ax.set_title('Two trigonometric Functions')

plt.show()
```

## Histogram

```python
#| fig-cap: 'Histogram'
#| fig-width: 3

plt.style.use('fivethirtyeight')
np.random.seed(1234)
x = np.random.chisquare(10, 100)

fig, ax = plt.subplots()

ax.hist(x, 30)
ax.set_xlabel('Bin range')
ax.set_ylabel('Frequency')
ax.set_title('Histogram')

fig.tight_layout()
plt.show()
```
